Understanding Features, Instances, and Their Relationships
* Features: These are the variables or attributes used for making predictions. In the Titanic dataset, features include age, fare, and passenger class.
* Instances: These are individual data points or records. Each row in the Titanic dataset is an instance representing a passenger.
* Relationships: Understanding how features relate to each other and the target variable (e.g., survival) is crucial for building effective models.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

data = {
    'Age': [25, 35, 45, 20, 30, 40, 50],
    'Salary': [50000, 60000, 70000, 40000, 55000, 65000, 75000]
}
df_ = pd.DataFrame(data)
df_
```

```python
stand_scalar = StandardScaler()
df_standard_scaled = df_.copy()
```

```python
df_standard_scaled[["Age", "Salary"]] = stand_scalar.fit_transform(df_[['Age', 'Salary']])
```

```python
minmax_scaler = MinMaxScaler()

df_minmax_scaled = df.copy()
df_minmax_scaled[['Age', 'Salary']] = minmax_scaler.fit_transform(df[['Age', 'Salary']])
```

```python
# How to handle the skewed and how we will handle this skewness with log transformation
print("skewness beofore the log transformation :", df["Fare"].skew())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = df.dropna(subset=["Fare"])
```

```python
# skewed data before log transformation


# Plotting the distribution and calculating skewness before transformation
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Fare'], kde=True)
plt.title('Fare Distribution Before Log Transformation')
```

```python
print("Skewness before log transformation:", df['Fare'].skew())
```

```python
# Log transformation
df['Fare'] = np.log1p(df['Fare'])
```

```python
#  Plotting the distribution and calculating skewness after transformation
plt.subplot(1, 2, 2)
sns.histplot(df['Fare'], kde=True)
plt.title('Fare Distribution After Log Transformation')

print("Skewness after log transformation:", df['Fare'].skew())

# Show the plots
plt.show()

# Print the first few rows of the dataframe to verify the transformation
print(df.head())
```
